### Stacks (Cтек)

A stack is a linear data structure that follows the **Last In, First Out (LIFO)** principle. This means that the last element added (pushed) to the stack will be the first one to be removed (popped). The stack has only one end for both inserting and removing elements, which is known as the top.

* Push/Append/Add (Add an element to the top): O(1) ✨
* Pop (Remove the top element): O(1)
* Peek (View the top element): O(1)

### Queues (Очередь)

A queue is a linear data structure that follows the **First In, First Out (FIFO) principle**. This means that the first element added to the queue will be the first one to be removed. A queue has two primary ends: the **front** (where elements are removed) and the **rear/end** (where elements are added).

* Enqueue (Add an element to the end): O(1)
* Dequeue (Remove an element from the front): O(1)
* Peek (View the front element): O(1)

### Tree (Дерево)


A tree is a **hierarchical data structure** made up of nodes (узлы), where each node has a value and links to child nodes. Trees have a top node called the root (корень) and are often visualized as an upside-down structure, with the root at the top and branches extending downward. Unlike linear data structures (like arrays or linked lists), trees allow for efficient organization and retrieval of hierarchical data, where each element can have multiple children.

**Key Characteristics:**

- Root: The top node, where traversal `(обход) of the tree begins.
- Node: Each element of a tree is called a node, consisting of a value and pointers to child nodes.
- Parent and Child Nodes: Each node can have zero or more child nodes, and each child node has exactly one parent.
- Leaf Node (лист): A node with no children.
- Depth and Height: The depth of a node is the distance from the root to that node, while the height of a tree is the longest path from the root to a leaf node.

**Types of the trees:**

1.	**Binary Tree** (Бинарное дерево): A tree where each node has at most two children (left and right).
2.	**Binary Search Tree (BST)** (Бинарное дерево поиска): A binary tree in which the left child’s value is less than its parent’s value, and the right child’s value is greater than the parent’s value.
3.	**Balanced Tree** (Сбалансированное дерево): A tree in which the heights of the left and right subtrees of any node differ by at most one.
4.	**AVL Tree**(АВЛ-дерево) and Red-Black Tree (Красно-черное дерево): Special balanced trees used to maintain a predictable structure for fast lookups, inserts, and deletes.
5.	**Heap (Куча)**: A complete binary tree that satisfies the heap property (max-heap or min-heap), where the parent node is either always greater or always smaller than its children.
6. **A Segment Tree** is a binary tree data structure used for efficient range queries and updates on an array. It is particularly useful when working with problems that involve operations over a range of elements (like sum, minimum, maximum, greatest common divisor, etc.).
    - Structure:
     - A binary tree where each node represents an interval (segment) of the array.
	 -  The root node represents the entire array, and each leaf node represents a single element.

### Graphs (Графы)

A graph is a data structure consisting of a set of vertices (also called **nodes** — узлы or вершины) and a set of **edges** (ребра) that connect pairs of vertices. Graphs are used to represent relationships between pairs of elements, making them useful in applications where entities are connected, such as social networks, transportation networks, or web page links.

Graphs are often represented visually as circles (vertices) connected by lines (edges), where each line represents a connection or relationship.


**Key Characteristics:**

- Directed Graph (Ориентированный граф): Each edge has a direction, meaning connections go one way, like following a one-way street.
- Undirected Graph (Неориентированный граф): Edges have no direction, so connections are bidirectional, like a two-way street.
- Weighted Graph (Взвешенный граф): Edges have weights, representing the cost or distance between vertices, such as distances between cities.
- Unweighted Graph (Невзвешенный граф): All edges are treated equally, with no associated weight.
- Cyclic and Acyclic Graphs (Циклический и ациклический графы): A cyclic graph contains at least one loop or cycle, while an acyclic graph has no cycles. A DAG (Directed Acyclic Graph) is commonly used in scheduling and dependency management.

### Clas Structure in Python

````python
class ClassName:
    """Description of the class"""

    # Class attribute (shared across all instances)
    class_variable = value

    def __init__(self, param1, param2):
        """Constructor method (initializes instance attributes)"""
        self.attribute1 = param1
        self.attribute2 = param2

    def instance_method(self):
        """Method that works on an instance's data"""
        return self.attribute1

    @classmethod
    def class_method(cls):
        """Method that works on the class as a whole"""
        return cls.class_variable

    @staticmethod
    def static_method():
        """Independent method within the class scope"""
        return "Static result"
````

### Class Method (Метод класса) VS Static Method (Статический метод)

Class Method (Метод класса)

- A method that operates on the class itself rather than individual instances.
- It can access and modify class attributes shared across all instances.
- Defined with the @classmethod decorator and uses cls as the first parameter.

Static Method (Статический метод)

- A method that does not operate on the instance or class.
- It is independent of both and is used for utility functions within the class context.
- Defined with the @staticmethod decorator and does not use self or cls.

### Decorator (Декоратор)

A decorator is a function that allows you to modify or extend the behavior of other functions or methods without changing their code directly. It is a form of metaprogramming (метапрограммирование), where you wrap one function with another to enhance or modify its functionality.

How Decorators Work

- A decorator takes a function as an argument, modifies or enhances it, and returns a new function.
- They are typically applied using the @decorator_name syntax above the function definition.

````python
def decorator_function(original_function):
    def wrapper_function():
        print("Before the function is called.")
        original_function()
        print("After the function is called.")
    return wrapper_function

@decorator_function  # Applying the decorator
def display():
    print("Display function.")

display()

# Output
# Before the function is called.
# Display function.
# After the function is called.
````

### Generator (Генератор)

A generator is a special type of iterator (итератор) that allows you to iterate over a sequence of values lazily, meaning that it generates values on the fly as you need them, rather than storing the entire sequence in memory. This can be particularly useful for working with large data sets or streams where memory efficiency is important.

```
def generator_function():
    yield value
```



````python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield the next value
        count += 1

counter = count_up_to(5)
for num in counter:
    print(num)

# 1
# 2
# 3
# 4
# 5
````

### Iterator (Итератор)

An iterator is an object that allows you to iterate over a sequence of values one at a time. In Python, an iterator must implement two methods: `__iter__()` and `__next__()`.

**Iterator Protocol**

Python defines an iterator protocol:
1.	An object is considered iterable if it implements the `__iter__()` method. This method should return an iterator.
2.	The iterator should implement the `__next__()` method to retrieve the next item in the sequence. If there are no more items, it should raise StopIteration.

How to implement iterator?


````python
# Using class

class Countdown:
    def __init__(self, start):
        self.start = start

    def __iter__(self):
        return self  # The iterator object itself

    def __next__(self):
        if self.start <= 0:
            raise StopIteration  # End of iteration
        self.start -= 1
        return self.start

countdown = Countdown(5)
for number in countdown:
    print(number)

# 5
# 4
# 3
# 2
# 1
# 0
````

````python
# Using generator

def countdown(start):
    while start >= 0:
        yield start
        start -= 1

for number in countdown(5):
    print(number)
````

### GIL (Global Interpreter Lock)

The Global Interpreter Lock (GIL) is a mechanism that allows only one thread to execute Python bytecode at a time, even on multi-core machines. It helps ensure memory safety but limits multi-threading performance, especially for CPU-heavy tasks.

**Key Points:**

1.	Single-threaded Execution: Even though Python supports threads, the GIL only allows one thread to run Python code at a time, limiting performance for CPU-bound tasks.
2.	Thread Safety: The GIL simplifies memory management by preventing multiple threads from modifying memory at the same time.
3.	Impact on Tasks:
- CPU-bound tasks: Performance is limited because only one thread can run at a time.
- I/O-bound tasks: Less of an impact, as threads can release the GIL when waiting for I/O, allowing other threads to run.

**How to work around GIL:**

- Multiprocessing: Use multiple processes (not threads), each with its own memory, to bypass the GIL for CPU-bound tasks.

### Machine Learning

In machine learning, tasks are typically categorized into supervised learning, unsupervised learning, and reinforcement learning. Below is a brief overview of the main components in a typical machine learning task:

**Key Components**

1.	**Features** (Фичи)
- Features are the input variables used to predict the target variable.
- In a dataset, features are usually represented as a matrix $X$, where each row represents a sample (data point), and each column represents a feature.

   **Feature Matrix** $X$:
$$
X = \begin{bmatrix}
x_1^{(1)} & x_2^{(1)} & \dots & x_n^{(1)} \
x_1^{(2)} & x_2^{(2)} & \dots & x_n^{(2)} \
\vdots & \vdots & \ddots & \vdots \
x_1^{(m)} & x_2^{(m)} & \dots & x_n^{(m)}
\end{bmatrix}
$$
- $m$ is the number of data points (samples).
- $n$ is the number of features (input variables).

2.	**Target** (Цель)
- The target (or label) is the output variable we aim to predict or classify. It is typically denoted as $y$ and can either be continuous (in regression tasks) or discrete (in classification tasks).
**Target Vector** $y$:
$$
y = \begin{bmatrix}
y^{(1)} \
y^{(2)} \
\vdots \
y^{(m)}
\end{bmatrix}
$$

3. **Machine Learning Tasks**
- Supervised Learning (Обучение с учителем): The model is trained on labeled data, where each input $X$ has a corresponding target $y$.
 - Regression (Регрессия): Predicts continuous values, e.g., predicting house prices.
 - Classification (Классификация): Predicts discrete labels, e.g., classifying emails as spam or not spam.
- Unsupervised Learning (Обучение без учителя): The model works with data that has no labels, trying to find patterns or structures within the data.
 - Clustering (Кластеризация): Groups data points into clusters, e.g., customer segmentation.
 - Dimensionality Reduction (Снижение размерности): Reduces the number of features, e.g., using Principal Component Analysis (PCA).
- Reinforcement Learning (Обучение с подкреплением): The agent learns to make decisions by interacting with the environment and receiving rewards or penalties.


Equation for a Supervised Learning Problem

In supervised learning, the goal is to learn a function $f$ that maps input features $X$ to the target $y$. The model’s prediction is often represented as:

$$
\hat{y} = f(X) + \epsilon
$$

Where:
- $\hat{y}$ is the predicted target.
- $X$ is the matrix of input features.
- $f(X)$ is the function (or model) that maps the input features to the output.
- $\epsilon$ represents the error (residual) between the predicted and true values.

### Overfitting VS Underfitting

1. **Overfitting:**.    

Definition: Overfitting happens when a model learns the details and noise in the training data to the extent that it negatively impacts its performance on new, unseen data (test data). In other words, the model becomes too complex, capturing patterns that are specific to the training data but do not generalize to new data.

Test and Training Error:
- Training Error: Low (because the model has learned the training data well, even capturing noise).
- Test Error: High (because the model is too complex and doesn’t generalize well to new data).
- In a graph of model complexity (x-axis) vs. error (y-axis), overfitting occurs when the model’s error on the test set is significantly higher than its error on the training set as the model becomes too complex.

2. **Underfitting:**.     

Definition: Underfitting occurs when a model is too simple to capture the underlying patterns in the data. This can happen when the model is not complex enough, either because it’s too simple or not trained long enough.

Test and Training Error:     
- Training Error: High (because the model is unable to learn the patterns in the training data).
- Test Error: High (since the model is too simple to generalize well to the test data).
- In the complexity vs. error graph, underfitting occurs when the model’s error on both the training set and test set remains high, even as model complexity increases.

### How to solve Overfitting/Underfitting

**Techniques to Overcome Overfitting:**.   

1.	**Cross-validation:**.   
Use k-fold cross-validation to evaluate the model’s performance across different subsets of the data. This helps to ensure that the model generalizes well across different parts of the data.

2.	**Regularization:**     
Regularization techniques like L1 (Lasso) and L2 (Ridge) regularization penalize the model for having large coefficients. This encourages the model to keep the weights small and simplifies the model, preventing it from fitting noise in the data.

3.	**Pruning (for decision trees):**      
Decision trees can overfit easily by growing too large. Pruning reduces the size of the tree by removing branches that provide little predictive power.

4.	**Ensemble Methods:**      
Methods like Bagging (e.g., Random Forest) and Boosting (e.g., Gradient Boosting) can help reduce overfitting. Bagging creates multiple models and averages their predictions, while boosting builds sequential models that correct previous errors.

5.	**Early Stopping (for neural networks):**     
During training, stop the model when the performance on the validation set starts to degrade, even if performance on the training set continues to improve. This prevents the model from learning noise and overfitting the training data.       

6.	**Dropout (for deep learning):**        
Dropout randomly drops units (neurons) during training, forcing the network to learn redundant representations. This reduces the likelihood of overfitting in neural networks.   

7.	**Reduce model complexity:**.          
Simplify the model by reducing the number of features or parameters, or by using simpler algorithms (e.g., linear models instead of highly complex models like deep neural networks).

---


**Techniques to Overcome Underfitting:**

1.	**Increase Model Complexity:**       
Use more complex models or algorithms. For example, if using linear regression, try using polynomial features or switch to more complex models like decision trees or neural networks.

2.	**Increase Training Time**:        
Train the model longer to allow it to learn more from the data. This is especially important in machine learning models like neural networks that require multiple epochs to learn effectively.

3.	**Feature Engineering:**.         
Add more features or better features. In regression tasks, try adding interaction terms or using domain-specific knowledge to create new features that better represent the data.

4.	**Decrease Regularization:**
If you’re using regularization techniques like Lasso (L1) or Ridge (L2), reduce the regularization strength (reduce the penalty on large coefficients), so the model has more freedom to learn the underlying patterns in the data.

5.	**Use More Complex Models:**.      
If a simpler model like linear regression isn’t capturing enough of the data, try a more complex model like random forests, SVM, or deep learning.

6.	**Increase Data Size:**.     
Sometimes underfitting can occur because the model doesn’t have enough data to learn from. Collecting more data can help the model learn better patterns.

7.	**Tune Hyperparameters:**.             
Adjust the model’s hyperparameters (e.g., depth of a decision tree, number of hidden layers in a neural network) to make the model more flexible and capable of capturing complex patterns.

## Regression

### Regression metrics

**1. Mean Absolute Error (MAE)**

$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$

Pros:
- Easy to interpret, as it represents the average magnitude of the errors.
- Less sensitive to outliers compared to MSE or RMSE.

Cons:
- It doesn’t penalize large errors as much as other metrics like MSE or RMSE.

---

**2. Mean Squared Error (MSE)**

$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

Pros:
- Penalizes larger errors more than MAE, making it useful when you want to prioritize minimizing larger deviations.
- Differentiable, which is useful in optimization problems.

Cons:
- Very sensitive to outliers, as large errors are squared, which may lead to overemphasis on them.

---

**3. Root Mean Squared Error (RMSE)**

$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$

Pros:
- Interpretable in the same units as the target variable, making it easier to understand in practical terms.
-Provides more weight to larger errors, similar to MSE.

Cons:
- Sensitive to outliers due to the squaring of errors.
- Can be misleading if there are large discrepancies in error distributions.

---

**4. R-squared (R²)**

$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}i)^2}{\sum{i=1}^{n} (y_i - \bar{y})^2}$

- $y_i$ is the actual value of the target variable (dependent variable) for the i-th data point in the dataset.
- $\hat{y}_i$ is the predicted value from your model for the i-th data point.
- $\bar{y}$ is the mean of the actual values of the target variable across all data points.


Pros:
- Provides a measure of how well the model explains the variance of the data, making it intuitive.
- Ranges from 0 to 1, where higher values indicate better fit.

Cons:
- Can be misleading, especially in cases of non-linear relationships.
- It can be artificially high with more predictors (even if they don’t improve the model significantly)


---

**5. Adjusted R-squared (Adjusted R²)**

$\text{Adjusted } R^2 = 1 - \left( \frac{1 - R^2}{n - 1} \right) - (n - p - 1)$

- n: The number of data points (or observations) in your dataset.
- p: The number of predictors (or independent variables) in the model.

Pros:
- Accounts for the number of predictors in the model, which helps in avoiding overfitting.
- More reliable for comparing models with different numbers of predictors.

Cons:
- Can still be misleading, especially with highly non-linear data.

---

**Mean Absolute Percentage Error (MAPE)**

$\text{MAPE} = \frac{100}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|$

Pros:
- Expressed as a percentage, which makes it easy to understand.
- Useful for relative error measurement across different scales.

Cons:
- Can be biased when actual values ($y_i$) are close to zero, leading to infinite or undefined values.
- Not ideal when errors are disproportionately high for small values of $y_i$.


### Regression Loss Functions

**1.	Mean Squared Error (MSE):**

Formula:        
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$
- n : Number of data points.
- $y_i$ : True value for data point  i .
- $\hat{y}_i$ : Predicted value for data point  i .
Description:       
Penalizes larger errors more heavily by squaring them, emphasizing outliers. Common for regression tasks.

---

2.	**Mean Absolute Error (MAE):**

Formula:     
$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$
- n : Number of data points.
- $y_i$ : True value for data point  i .
- $\hat{y}_i$ : Predicted value for data point  i .
Description:        
Averages the absolute differences between predicted and actual values, treating all errors equally.

### Regression models

**1. Multiple Linear Regression**.     

Description:       
Extends simple linear regression to handle multiple features:
$$y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p + \epsilon$$
- $X_1, X_2, \dots, X_p$: Features.
- $\beta_1, \beta_2, \dots, \beta_p$: Coefficients for the features.

Benefits:
- Can model relationships with multiple features.
- Still interpretable, particularly with fewer features.

---


2. **Ridge Regression (L2 Regularization)**.  

Description:     

Adds a penalty term to the loss function to shrink large coefficients:           
Loss function:       

$$L_{\text{Ridge}} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}i)^2 + \lambda \sum_{j=1}^{p} \beta_j^2$$




---

3. **Lasso Regression (L1 Regularization)**

Description:     

Adds a penalty term based on the absolute values of coefficients:       
Loss function:      
$$
L_{\text{Lasso}} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}i)^2 + \lambda \sum_{j=1}^{p} |\beta_j|
$$



Benefits:
- Performs feature selection by driving some coefficients to exactly zero.
- Suitable for high-dimensional data.

## Classification

### TP, TN, FP, FN

- **True Positives (TP)**:   
  The number of instances that were correctly predicted as positive. In other words, the model predicted “positive” and the actual label was also “positive.”
- **True Negatives (TN)**:
  The number of instances that were correctly predicted as negative. The model predicted “negative” and the actual label was also “negative.”
- **False Positives (FP)**:
  The number of instances that were incorrectly predicted as positive. The model predicted “positive,” but the actual label was “negative.” This is also called a **“Type I error.”**
- **False Negatives (FN)**:
  The number of instances that were incorrectly predicted as negative. The model predicted “negative,” but the actual label was “positive.” This is also called a **“Type II error.”**


\begin{bmatrix}
\text{True Positives (TP)} & \text{False Positives (FP)} \\
\text{False Negatives (FN)} & \text{True Negatives (TN)}
\end{bmatrix}


### Classification metrics

1. **Accuracy**.        
Formula:     
$\text{Accuracy} = \frac{\text{True Positives} + \text{True Negatives}}{\text{Total Number of Instances}}$.    

Pros:
- Simple to understand and compute.
- Suitable for balanced datasets.    

Cons:    
- Can be misleading for imbalanced datasets, where the majority class dominates.


---


2. **Precision**

Formula:     
$\text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}$

Pros:     
- Focuses on the relevance of the positive predictions.
- Useful when false positives are costly or undesirable.     

Cons:
- Doesn’t consider false negatives, so it may overlook the model’s ability to correctly identify all actual positive instances.

---


3. **Recall (Sensitivity/TPR)**

Formula:      
$\text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}$.       

Pros:
- Focuses on the model’s ability to correctly identify all positive instances.
- Useful when false negatives are costly or undesirable.

Cons:
- Doesn’t account for false positives, which could lead to an overly optimistic view of model performance.

---

4. **F1-score**

Formula:      
$\text{F1} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$.    

Pros:       
- Balances Precision and Recall, providing a single metric for both.
- Suitable for imbalanced datasets where both false positives and false negatives are important.

Cons:     
- May not provide a clear picture of the model’s performance if the Precision and Recall are vastly different.


---


5. **Area Under the ROC Curve (AUC-ROC)**

Formula:     
$\text{AUC} = \int_{0}^{1} \text{True Positive Rate (Recall)} , \text{d}(\text{False Positive Rate})$.     

Pros:     
- Provides an aggregate measure of performance across all classification thresholds.
-  Can be used to compare classifiers and doesn’t depend on a specific decision threshold.    

Cons:     
- May not be as interpretable as other metrics like Accuracy, Precision, or Recall in specific contexts.

---


6. **Log Loss (Cross-Entropy Loss)**.      


Log Loss (Логарифмическая ошибка, Кросс-энтропия) is a metric that quantifies the error of a probabilistic prediction model. It evaluates how well the predicted probability distributions align with the actual class labels.         

Formula:         
For binary classification:          
$$
\text{Log Loss} = - \frac{1}{n} \sum_{i=1}^{n} \left( y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right)
$$

Where:           
- n : Number of data points.
- $y_i$ : True label ( 0  or  1 ).
- $\hat{y}_i$ : Predicted probability of the positive class ($ 0 \leq \hat{y}_i \leq 1 $).

For multiclass classification:         
$$
\text{Log Loss} = - \frac{1}{n} \sum_{i=1}^{n} \sum_{j=1}^{k} y_{ij} \log(\hat{y}_{ij})
$$

Where:
- k : Number of classes.
- $y_{ij}$ : True label (1 if the instance belongs to class  j , 0 otherwise).
- $\hat{y}_{ij}$ : Predicted probability of class  j .\

Key Characteristics:    

- Range: Always non-negative ( \text{Log Loss} \geq 0 ). Smaller values indicate better performance.
- Focus on Probabilities: Unlike metrics like accuracy, Log Loss penalizes models that assign incorrect probabilities to predictions. Overconfidence in wrong predictions leads to higher Log Loss.

###  ROC and AUC

True Positive Rate (TPR):
$$TPR = \frac{TP}{TP + FN}$$.   

False Positive Rate (FPR):
$$FPR = \frac{FP}{FP + TN}$$.

The ROC curve is a visual representation of model performance across all thresholds.     

The ROC curve is drawn by calculating the true positive rate (TPR) and false positive rate (FPR) at every possible threshold (in practice, at selected intervals), then graphing TPR over FPR. For a binary classifier, a model that does exactly as well as random guesses or coin flips has a ROC that is a diagonal line from (0,0) to (1,1). The AUC is 0.5, representing a 50% probability of correctly ranking a random positive and negative example.

### Classification Loss Functions


1.	**Binary Cross-Entropy (Log Loss):**.     

Formula:       
$$L = - \frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$$
- n : Number of data points.
- $y_i$ : True label ( 0  or  1 ).
- $\hat{y}_i$ : Predicted probability for the positive class ( 0 \leq \hat{y}_i \leq 1 ).

Description:        
Measures how well the predicted probabilities match the actual labels. Common for binary classification.


---

2.	**Categorical Cross-Entropy (Multiclass Log Loss):**

Formula:       
$$L = - \frac{1}{n} \sum_{i=1}^{n} \sum_{j=1}^{k} y_{ij} \log(\hat{y}_{ij})$$.

- n : Number of data points.    
- k : Number of classes.       
$y_{ij}$ : True label (1 if class  j  is the true class for instance  i , otherwise 0).        
- $\hat{y}_{ij}$ : Predicted probability for class  j  for instance  i .

Description:      
Extends binary cross-entropy to handle multiple classes. Suitable for multiclass classification tasks.

### Ensemble Methods

Definition:
Ensemble methods are techniques that combine multiple models (often weak models) to improve performance. The idea is that a group of models working together can give better predictions than any single model alone.

Types:      
- **Bagging** (Bootstrap Aggregating): It involves training multiple models in parallel on different subsets of the data and averaging the predictions (e.g., Random Forest).
- **Boosting**: It involves training models sequentially, where each new model tries to correct the mistakes of the previous ones (e.g., Gradient Boosting, AdaBoost).
- **Stacking**: Combining predictions from multiple models (which can be of different types), and using another model (called a meta-learner) to make the final prediction.

### Classification models (Intuition)

1.	**Logistic Regression**:     
Intuitive explanation: It tries to find the best line (or curve) that separates different classes. This line is used to predict the probability that a given input belongs to a certain class. It’s like drawing a dividing line and saying, “If data falls on one side, it’s class A; if it falls on the other side, it’s class B.”

2.	**K-Nearest Neighbors** (KNN):        
Intuitive explanation: Imagine you have a bunch of labeled points (data) on a graph. When you need to classify a new point, you look at its nearest neighbors (the closest points) and choose the most common class among them. It’s like asking your closest friends what they think about something, and going with the majority opinion.

3.	**Support Vector Machine** (SVM):     
Intuitive explanation: SVM tries to find the best boundary (called a “hyperplane”) that separates data into classes. It works by maximizing the distance between the boundary and the closest points from either class. Think of it as trying to put a fence between two groups of cows to keep them separated as much as possible.

4.	**Naive Bayes**:      
Intuitive explanation: It’s based on probability and assumes that features (predictors) are independent of each other (hence the “naive”). It calculates the likelihood of an item belonging to a class based on the features it has and the probability of those features in each class. It’s like trying to predict if someone likes a movie based on individual preferences (action, romance, comedy) and the probability of liking each genre.      

5.	**Decision Tree**:       
Intuitive explanation: A decision tree makes decisions by asking a series of yes/no questions. The data is split at each question based on a feature, and the process continues until the data reaches a “leaf,” which gives the final classification. It’s like a flowchart where you answer questions to classify an item.

6.	**Random Forest (based on bagging)**:         
Intuitive explanation: A random forest is a collection of many decision trees. Each tree makes its own prediction, and the final classification is determined by majority voting. It’s like asking many people for their opinion and taking the one that most of them agree on.

7.	**Gradient Boosting (based on boosting)** (and variants like XGBoost, LightGBM):     
Intuitive explanation: Gradient boosting builds decision trees one by one, where each new tree tries to correct the errors of the previous one. It’s like a group of people working together to solve a problem: the first person might make a mistake, but the next person learns from it and improves on the solution.

### Classification models (in details)

1. **Logistic Regression**      

Training:      
Logistic regression is trained using optimization techniques like Gradient Descent. The goal is to minimize the log-loss or cross-entropy loss function.         
$$ \text{Log-Loss} = -\frac{1}{N} \sum_{i=1}^N \left( y_i \log(p_i) + (1 - y_i) \log(1 - p_i) \right) $$
where:          
- $y_i$  is the actual label (0 or 1)
- $p_i$  is the predicted probability for class 1.         

Loss Function: The Log-Loss (cross-entropy) function measures the accuracy of the model’s predictions. The closer the predicted probability is to the actual label, the lower the log-loss.     

Interview Facts:        
- It is used for binary classification problems.
- It outputs probabilities and can be used for multi-class classification with a technique like One-vs-All.
- Assumption: Features are independent (for binary logistic regression).


---

2. **K-Nearest Neighbors (KNN)**

Training:        
KNN is a non-parametric algorithm, meaning it doesn’t require explicit training. It simply stores the training data and makes predictions based on it.

Loss Function:         
There isn’t a specific loss function used during training, as KNN directly classifies based on the distance between data points.

Prediction Process:       
- Given a new point, KNN finds the ‘K’ nearest neighbors using a distance metric like Euclidean distance.
- The class label of the majority of neighbors determines the classification.

Interview Facts:       
- KNN can be computationally expensive during prediction, as it needs to calculate the distance between the new point and all training points.
- The optimal value of  K  is determined using techniques like cross-validation.
- Works well for small datasets but is not scalable for large datasets.


---


3. **Support Vector Machine (SVM)**

Training:      
SVM is trained by finding the hyperplane that maximizes the margin between classes.

The hinge loss function is typically used:       
$$ L(\mathbf{w}, b) = \sum_{i=1}^N \max(0, 1 - y_i(\mathbf{w} \cdot \mathbf{x}_i + b)) $$
where:      
- $\mathbf{w}$  is the weight vector
- b  is the bias term
- $\mathbf{x}_i$  is the feature vector
- $y_i$  is the true class label (either 1 or -1)

Loss Function: The hinge loss is used for classification tasks. It penalizes points that are on the wrong side of the margin.     

Interview Facts:    
- SVM works well in high-dimensional spaces and for cases where the margin of separation is clear.
- It can use kernels (like polynomial or RBF) to handle non-linear classification tasks.
- SVM is sensitive to the choice of kernel and the values of hyperparameters like  C  (regularization) and  \gamma  (kernel parameter)


---


4. **Decision Tree**

Training: Decision trees use a greedy algorithm to build the tree by selecting the best feature to split on. The feature is selected based on a metric like Gini Impurity or Information Gain.       

Gini Impurity formula:            
$$ Gini = 1 - \sum_{i=1}^k p_i^2 $$
where  $p_i$  is the probability of samples belonging to class  at a given node.
Entropy (for Information Gain):     
$$ Entropy = - \sum_{i=1}^k p_i \log_2(p_i) $$.     

Loss Function: Decision trees use Gini impurity or entropy to split nodes and minimize classification error.      

Interview Facts:        
- Decision trees are prone to overfitting. Pruning techniques like pre-pruning (limiting tree depth) and post-pruning are used to control overfitting.
- They are interpretable and can handle both categorical and continuous variables.


---

5. **Random Forest**

Training: Random Forest is an ensemble method that builds multiple decision trees using Bootstrap Aggregating (bagging). Each tree is trained on a random subset of the data with replacement.

- Node Selection: When creating each decision tree, only a random subset of features is considered for splitting at each node (rather than all features). This helps reduce correlation between the trees and makes the forest more robust.
- Prediction: For classification, the final prediction is made by majority voting: each tree predicts a class, and the class with the most votes is selected as the final result.

Loss Function: Random Forest doesn’t use a direct loss function. It uses the decision trees’ splitting criteria (Gini Impurity or Information Gain).

Interview Facts:     
- Random Forest can handle overfitting better than individual decision trees.
- Works well for large datasets and can handle missing values.
- Out-of-Bag (OOB) error estimation can be used for model validation.


---


6. **Gradient Boosting (and variants like XGBoost)**

Training: Gradient Boosting builds trees sequentially. Each tree corrects the errors (residuals) made by the previous tree. It minimizes a loss function by iteratively improving predictions.        

Loss Function: Gradient Boosting minimizes a specific loss function (like mean squared error for regression or log-loss for classification) through gradient descent.       

$$ L = \sum_{i=1}^N \left[ y_i - \hat{y}_i \right]^2 $$  for regression tasks.
    
$$ L = -\sum_{i=1}^N \left( y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right) $$  for binary classification.     


Prediction: The model makes predictions by combining the predictions of all the individual trees, where each tree contributes more to correcting the mistakes of previous trees.      

Interview Facts:     
- Gradient Boosting is a sequential method, meaning trees are added one by one to correct previous mistakes.
- It’s highly sensitive to hyperparameters like learning rate, number of estimators, and depth of trees.
- XGBoost and LightGBM are optimized implementations of Gradient Boosting that are faster and more efficient.

# Statistics

### Law of Large Numbers (Закон Больших Чисел)

The Law of Large Numbers states that as the size of a sample from a population increases, the sample mean (average) will get closer and closer to the true population mean. In other words, the more data you collect, the more reliable your estimate of the average will be.

### Central Limit Theorem (Центральная предельная теорема)

It states that if you take many random samples from any population, regardless of the original distribution of the data, the distribution of the sample means will tend to be normal (i.e., bell-shaped), provided the sample size is large enough.


---

If we have a population with:
- Mean = μ (Математическое ожидание)
- Standard deviation = σ (Стандартное отклонение)

Then, if we take n random samples, the sampling distribution of the sample mean will approach a normal distribution with:
- Mean = μ (same as the population mean)
- Standard deviation = σ / √n (this is smaller than the population standard deviation).       

This is true for large n. The larger the sample size, the closer the sampling distribution will be to a normal distribution.

### Bayes’ Theorem (Теорема Байеса)

**Intuition**:      
- You have some prior knowledge or belief about the probability of an event happening.
- Then you get new data or evidence.
- Bayes’ Theorem helps you combine both: the prior belief and the new evidence, to arrive at a more accurate posterior belief or updated probability.

**Formula**:        

The formula for Bayes’ Theorem is:     

$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$.       

Where:     
- P(A|B)  is the posterior probability — the probability of event A occurring given that B has occurred.
- P(B|A)  is the likelihood — the probability of event B occurring given that A has occurred.
- P(A)  is the prior probability — the initial probability of event A before we have any new data.
- P(B)  is the evidence — the total probability of event B occurring (regardless of A).

# Data Engineering

### ETL

**1. Extract (Извлечение)**
  
In this stage, data is gathered from various source systems. These sources could include databases, APIs, flat files (CSV, Excel), cloud storage, or even web scraping.      

- Goal: Collect raw data from multiple sources without altering it too much.
- Example: Extracting sales data from a MySQL database, customer information from a CRM, or transactional data from an API.

---

**2. Transform (Преобразование)**

Once the data is extracted, it needs to be processed and transformed to ensure it is in the correct format, cleaned, and enriched before loading it into the target system.

Transformation steps can include:      
- Cleaning: Removing duplicates, handling missing values, correcting data errors.
- Filtering: Removing irrelevant data.
- Aggregating: Summing or averaging data.
- Normalization: Ensuring data is in a consistent format (e.g., dates, currencies).
- Joining: Combining data from multiple sources into a single dataset.    

Example: Calculating the total sales for a region or converting text-based dates into a standard date format.

---


**3. Load (Загрузка)**

This stage involves transferring the transformed data into the destination system, typically a data warehouse or data lake.     

Goal: Ensure data is loaded efficiently and is ready for analysis or reporting.      

Types of loading:      
- Full Load: Replacing all the data in the destination with the new data.
- Incremental Load: Only loading new or changed data to avoid reloading the entire dataset.

Example: Loading customer sales data into a data warehouse where business analysts can query and report on it.

### ETL VS ELT

In addition to ETL, there is also the ELT (Extract, Load, Transform) process. The key difference is that in ELT, the data is loaded into the target system first, and the transformation happens afterward.    

- ETL: Data is transformed before loading it into the data warehouse.
- ELT: Data is loaded before transforming it.        

ETL is more suitable when data transformation is complex or requires a heavy computational load, while ELT is used when the target system (e.g., modern cloud data warehouses like Google BigQuery, Snowflake) has sufficient processing power to handle transformations efficiently.

### Data Orchestration (Операции с данными)


Data orchestration involves managing the flow of data through various stages in a pipeline, ensuring that tasks are executed in the correct order, with proper scheduling, dependencies, and error handling.

- Used for: Coordinating complex data workflows and processes.
- Example: Scheduling ETL jobs, ensuring tasks run sequentially or in parallel as needed.
- Tools: Apache Airflow, Luigi, Dagster.
